```python
import pandas as pd
from google.colab import files
```

```python
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
print(f'Uploaded file: {file_name}')
```

```python
try:
    df = pd.read_excel(file_name)
    print(f"File '{'census_2011.xlsx'}' loaded successfully.")
except Exception as e:
    print(f"Error loading file: {e}")
```

```python
df.rename(columns={
    'State name': 'State/UT',
    'District name': 'District',
    'Male_Literate': 'Literate_Male',
    'Female_Literate': 'Literate_Female',
    'Rural_Households': 'Households_Rural',
    'Urban_Households': 'Households_Urban',
    'Age_Group_0_29': 'Young_and_Adult',
    'Age_Group_30_49': 'Middle_Aged',
    'Age_Group_50': 'Senior_Citizen',
    'Age not stated': 'Age_Not_Stated'
}, inplace=True)
print(df.head())
```

```python
def state(name):
    words = name.split()
    return ' '.join([word.capitalize() if word.lower() != 'and' else word.lower() for word in words])

df['State/UT'] = df['State/UT'].apply(state)
print(df['State/UT'].unique())
```

```python
telangana = ['Adilabad', 'Hyderabad', 'Karimnagar', 'Khammam', 'Mahbubnagar', 'Medak', 'Nalgonda', 'Nizamabad', 'Rangareddy', 'Warangal']
ladakh = ['Leh', 'Kargil']

df.loc[df['District'].isin(telangana), 'State/UT'] = 'Telangana'
df.loc[df['District'].isin(ladakh), 'State/UT'] = 'Ladakh'

print(df['State/UT'].unique())
```

```python
data_before = df.isnull().mean() * 100
print("Missing data before filling:\n", data_before)

if 'Population' in df.columns and 'Male' in df.columns and 'Female' in df.columns:
    df['Population'].fillna(df['Male'] + df['Female'], inplace=True)
if 'Literate' in df.columns and 'Literate_Male' in df.columns and 'Literate_Female' in df.columns:
    df['Literate'].fillna(df['Literate_Male'] + df['Literate_Female'], inplace=True)
if 'Population' in df.columns and 'Young_and_Adult' in df.columns and 'Middle_Aged' in df.columns and 'Senior_Citizen' in df.columns and 'Age_Not_Stated' in df.columns:
    df['Population'].fillna(df['Young_and_Adult'] + df['Middle_Aged'] + df['Senior_Citizen'] + df['Age_Not_Stated'], inplace=True)
if 'Households' in df.columns and 'Households_Rural' in df.columns and 'Households_Urban' in df.columns:
    df['Households'].fillna(df['Households_Rural'] + df['Households_Urban'], inplace=True)

data_after = df.isnull().mean() * 100
print("Missing data after filling:\n", data_after)
```

```python
import pandas as pd
from pymongo import MongoClient
import sqlite3

client = MongoClient("mongodb+srv://sachitananda98:1234@cluster0.l5k86c1.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0")
db = client['census_db']
collection = db['census']

data_dict = df.to_dict("records")
collection.insert_many(data_dict)

print("Data inserted into MongoDB successfully.")

data_from_mongo = list(collection.find({}, {'_id': 0}))

conn = sqlite3.connect('census_db.sqlite')
cur = conn.cursor()

def map_dtype(dtype):
    if pd.api.types.is_integer_dtype(dtype):
        return "INTEGER"
    elif pd.api.types.is_float_dtype(dtype):
        return "REAL"
    else:
        return "TEXT"

columns = df.columns.tolist()
column_definitions = ", ".join([f"'{col.replace(' ', '_').replace('/', '_')}' {map_dtype(dtype)}" for col, dtype in zip(columns, df.dtypes)])
create_table_query = f"CREATE TABLE census ({column_definitions});"

cur.execute("DROP TABLE IF EXISTS census")
cur.execute(create_table_query)

for record in data_from_mongo:
    values = tuple(record.values())
    insert_query = f"INSERT INTO census ({', '.join([col.replace(' ', '_').replace('/', '_') for col in columns])}) VALUES ({', '.join(['?' for _ in values])})"
    cur.execute(insert_query, values)

conn.commit()
conn.close()

print("Data uploaded to the relational database successfully.")
```
